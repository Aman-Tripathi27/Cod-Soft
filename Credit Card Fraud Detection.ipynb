### Seprating the data for analysis

```python
Normal = cc[cc.Class == 0]
Fraud = cc[cc.Class == 1]
(Normal.shape),(Fraud.shape)
```

```python
# Statistical measures of the data 
Normal.Amount.describe()
```

```python
Fraud.Amount.describe()
```

```python
#Comparing both transactions
cc.groupby('Class').mean()
```

### Concatenating 2 DataFrames

```python
Normal_sample = Normal.sample(492)
```

```python
ccd = pd.concat([Normal_sample, Fraud], axis=0)
ccd.head()
```

```python
ccd.tail()
```

```python
ccd['Class'].value_counts()
```

```python
ccd.groupby('Class').mean()
```

**Splitting the data**

```python
X = ccd.drop("Class", axis=1)
y = ccd["Class"]
```

```python
X
```

**comparing both transactions**

```python
y
```

**Training and testing of data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

### Accuracy score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
training_data_accuracy
```

```python
X_test_prediction = model.predict(X_test)
training_data_accuracy = accuracy_score(X_test_prediction, y_test)
```
